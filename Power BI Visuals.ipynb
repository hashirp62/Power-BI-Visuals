Q3.

 Steps to create a Line Chart showing the Sales Trend over time (by Month and Year).

 *NB: Always check the required and coulmn and clean it if needed before loading it into Power BI report view.*

 **1.** Open data view.

 **2.** create month year column

 This method is used to display sales data in a readable month and year format for better trend analysis in the line chart.

 - Go to Table Tools > New Column

 - Enter the formula:

   Month Year = FORMAT('Global_Superstore2 - Global_Sup'[Order Date], "MMM YYYY")

   Press ENTER.

**3.** Create MonthSort Column

This column is created to create a numeric values for each month and year, which helps power BI to display the trend right order.

- Again click Table Tools > New Column

- Enter:

  MonthSort = YEAR('Global_Superstore2 - Global_Sup'[Order Date])*100 + MONTH('Global_Superstore2 - Global_Sup'[Order Date])

Press ENTER

**4.** Sort the Month

Tells Power BI to arrange month year using the monthSort numbers.

- Select the Month Year column.

- Go to Column Tools > Sort by Column > MonthSort

**5.** Create Line Chart

- Go to report view and select line chart from visualization  pane.

**6.** Add fields

- drag:

  Month year > X-axis

  sales > Y-axis

**7(optional).** Make It Look Clean

-  Format and design for better readability and makes it easier to identify trends.

*Trend: From the chart we can see a higher sales repeatedly happening between the month of november and december every year.*
